# Analyst Command Center - Preliminary Data Exploration 🧪

This notebook demonstrates the exploratory data analysis (EDA) process used to validate the synthetic financial dataset. We focus on distribution, seasonality, and outlier detection.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Setting aesthetic theme
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = [12, 6]

# Load processed dataset
try:
    df = pd.read_csv('../data/processed/synthetic_financials.csv')
    df['date'] = pd.to_datetime(df['date'])
    print(f"Dataset loaded: {df.shape[0]} rows and {df.shape[1]} columns.")
except FileNotFoundError:
    print("Dataset missing. Generating on-the-fly...")
    import sys
    sys.path.append('..')
    from utils import load_and_generate_data
    df = load_and_generate_data()

df.head()
```

## 1. Portfolio Revenue Distribution
Understanding the spread of revenue across different companies helps identify market leaders.

```python
plt.figure(figsize=(10,6))
sns.histplot(df['revenue'], kde=True, color='skyblue')
plt.title('Distribution of Monthly Revenue across Portfolio')
plt.xlabel('Revenue ($)')
plt.show()
```

## 2. Sectoral Performance Analysis
Comparing gross margins across sectors to identify high-efficiency industries.

```python
df['gross_margin'] = df['gross_profit'] / df['revenue']
sns.boxplot(data=df, x='sector', y='gross_margin', palette='viridis')
plt.title('Gross Margin Variances by Sector')
plt.ylabel('Gross Margin (%)')
plt.show()
```

## 3. Correlation Heatmap
Checking for strong linear relationships between marketing spend, churn (implied), and revenue.

```python
corr = df[['revenue', 'marketing_spend', 'cac', 'customers', 'net_income']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Key Metric Correlation Matrix')
plt.show()
```
